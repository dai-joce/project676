```python
# Install libraries as needed (uncomment if running on a clean environment)
%pip install networkx pandas numpy matplotlib scipy scikit-learn
%pip install python-louvain igraph

import os, io, gzip, zipfile, tarfile, sys, math, random
import numpy as np
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt

# For reproducibility
random.seed(42)
np.random.seed(42)

print(nx.__version__)
```

```python
# === You may using the follow method to download datasets if you like ===
DATA_URL = "https://mcauleylab.ucsd.edu/public_datasets/data/amazon/qa/qa_Clothing_Shoes_and_Jewelry.json.gz"
LOCAL_PATH = "/content/data/qa_Clothing_Shoes_and_Jewelry.json.gz"

os.makedirs("data", exist_ok=True)

def download_dataset(url: str, to_path: str):
    import urllib.request
    print(f"Downloading from {url} ...")
    urllib.request.urlretrieve(url, to_path)
    size = os.path.getsize(to_path) / (1024*1024)
    print(f"Saved to {to_path} ({size:.2f} MB)")

# Uncomment to download when ready
download_dataset(DATA_URL, LOCAL_PATH)
```

```python
import gzip
import ast
import pandas as pd

FILE_PATH = '/content/data/qa_Clothing_Shoes_and_Jewelry.json.gz'

data = []
with gzip.open(FILE_PATH, 'rt', encoding='utf-8') as f:
    for i, line in enumerate(f):
        try:
            item = ast.literal_eval(line)
            data.append(item)
        except Exception as e:
            print("Error on line", i, ":", e)
        # REMOVE or comment out this line:
        # if i >= 9: break


df = pd.DataFrame(data)
print(df.shape)      # This will tell you how many rows were loaded
print(df.head())     # Preview first 5 rows
```

```python
# Assume you have: df = pd.DataFrame(data)

import matplotlib.pyplot as plt
import seaborn as sns

# 1. Distribution of Question Types
print("Question Type distribution:")
print(df['questionType'].value_counts())
df['questionType'].value_counts().plot(kind='bar')
plt.title('Distribution of Question Types')
plt.xlabel('Question Type')
plt.ylabel('Count')
plt.show()

# 2. Frequency of Top Items (Products)
top_asins = df['asin'].value_counts().head(10)
print("\nTop 10 most-asked-about ASINs:")
print(top_asins)
top_asins.plot(kind='bar')
plt.title('Top 10 Most Asked-About Products')
plt.xlabel('ASIN')
plt.ylabel('Number of Questions')
plt.show()

# 3. Distribution of Answer Types (for yes/no)
print("\nAnswer Type distribution (only for yes/no):")
df_yesno = df[df['questionType'] == 'yes/no']
print(df_yesno['answerType'].value_counts())
df_yesno['answerType'].value_counts().plot(kind='bar', color=['green','red','gray'])
plt.title('Distribution of Answer Types (Yes/No Questions)')
plt.xlabel('Answer Type')
plt.ylabel('Count')
plt.show()

# 4. Temporal gaps between responses (answerTime/unixTime)
df['answer_dt'] = pd.to_datetime(df['unixTime'], unit='s')
print("\nEarliest and latest answer dates:")
print(df['answer_dt'].min(), "-", df['answer_dt'].max())
df['answer_dt'].hist(bins=30)
plt.title('Distribution of Answer Dates')
plt.xlabel('Answer Date')
plt.ylabel('Number of Answers')
plt.show()

# 5. Most common words/phrases in questions
from collections import Counter
import re

# Lowercase and tokenize questions
words = []
for q in df['question']:
    words.extend(re.findall(r'\w+', q.lower()))
word_counts = Counter(words)
most_common_words = word_counts.most_common(15)
print("\nMost common question words:")
for word, count in most_common_words:
    print(f"{word}: {count}")
```
